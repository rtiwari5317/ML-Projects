```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
df = pd.read_csv('haberman.csv')
df.shape
```

```python
col_name = ['Age','Oprn_Year','Nodes Detected','Status']
df.columns = col_name
```

```python
df['Status'].value_counts()
```

```python
#Using Histogram Univariate Analysis with single attribute here 'Status'

df.Age.hist(grid = True,edgecolor='white', linewidth=1.5)
plt.show()
```

```python
df.plot(kind ='scatter',x ='Age', y ='Nodes Detected')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(df,hue = 'Status', size = 6)\
   .map(plt.scatter,'Age','Nodes Detected')\
   .add_legend()

plt.show()
```

```python
#Using Pair Plot
sns.set_style('whitegrid')
sns.pairplot(df,hue = 'Status',vars = ['Age','Oprn_Year','Nodes Detected'],size = 3)

plt.show()
```

```python
#Univariate Analysis Using HistoGram and PDF
sns.FacetGrid(df,hue = 'Status',size = 6 )\
   .map(sns.distplot,'Age')\
   .add_legend()

plt.show()

sns.FacetGrid(df,hue = 'Status',size = 6 )\
   .map(sns.distplot,'Oprn_Year')\
   .add_legend()

plt.show()

sns.FacetGrid(df,hue = 'Status',size = 6 )\
   .map(sns.distplot,'Nodes Detected')\
   .add_legend()

plt.show()
```
